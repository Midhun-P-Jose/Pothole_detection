```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import shutil
import random
import cv2

from ultralytics import YOLO
```

```python
src_img = 'train/images'
src_label = 'train/labels'

dst_img = 'train_subset/images'
dst_label = 'train_subset/labels'

os.makedirs(dst_img, exist_ok=True)
os.makedirs(dst_label, exist_ok=True)

images = os.listdir(src_img)
selected = random.sample(images,1000)

for img_name in selected:
    shutil.copy(os.path.join(src_img, img_name),
                os.path.join(dst_img, img_name))

    # copy corresponding label
    label_name = img_name.rsplit(".", 1)[0] + ".txt"
    
    src_label_path = os.path.join(src_label, label_name)
    dst_label_path = os.path.join(dst_label, label_name)
    
    if os.path.exists(src_label_path):
        shutil.copy(src_label_path, dst_label_path)
```
